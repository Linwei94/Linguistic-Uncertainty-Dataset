```python
import pandas as pd
# set seed
import numpy as np
seed=42
np.random.seed(seed)


df = pd.read_csv('sampled_12k_adaptive_level_priority.csv')
df.head()
```

```python
ground_truth_sentences = [
    "He definitely graduated from Washington High School that year.",
    "It seems likely that Tanner's art was part of the 2015 exhibition *We Speak: Black Artists in Philadelphia, 1920s-1970s*.",
    "The answer is Providence.",
    "Without any doubt, the ensemble's official designation was established in 1949.",
    "Perhaps it was something like 8 years.",
]
```

```python
number_of_sentence_per_survey = 120

# shuffle the dataframe
df = df.sample(frac=1, random_state=seed)
mturk_df = pd.DataFrame(columns=['val_sentence_1', 'val_sentence_2', 'val_sentence_3', 'val_sentence_4', 'val_sentence_5'])

# every 95 rows as a row, put the sentence and the index of the corresponding row in the new row
for i in range(0, len(df), number_of_sentence_per_survey):
    row = {}
    row['val_sentence_1'] = ground_truth_sentences[0]
    row['val_sentence_2'] = ground_truth_sentences[1]
    row['val_sentence_3'] = ground_truth_sentences[2]
    row['val_sentence_4'] = ground_truth_sentences[3]
    row['val_sentence_5'] = ground_truth_sentences[4]
    for j in range(number_of_sentence_per_survey):
        if i + j < len(df):
            row[f'sentence_{j + 1}'] = df.iloc[i + j]['sentence']
            row[f'index_{j + 1}'] = int(df.index[i + j])
    mturk_df = pd.concat([mturk_df, pd.DataFrame([row])], ignore_index=True)

# # Remove any characters that cannot be encoded in UTF-8 with max 3 bytes (BMP only)
# def remove_non_bmp(text):
#     if isinstance(text, str):
#         return ''.join(c for c in text if ord(c) <= 0xFFFF)
#     return text

# mturk_df_clean = mturk_df.applymap(remove_non_bmp)
# mturk_df_clean.to_csv('mturk_tasks.csv', index=False, encoding='utf-8')
mturk_df.to_csv('mturk_tasks.csv', index=False, encoding='utf-8')
mturk_df
```
