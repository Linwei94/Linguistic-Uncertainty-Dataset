```python
import pandas as pd

df = pd.read_csv("llm_uncertainty_results.csv")
df.head()
```

```python
confidence_method = "vanilla_uncertainty"

import matplotlib.pyplot as plt
from metrics import ECE
from sklearn.metrics import roc_auc_score
# Filter out rows where grader_result is 'C'
filtered_df = df[df[f"grader_result_{confidence_method}"] != "C"]


ece_score = ECE(n_bins=15)

confidence_scores = filtered_df[f"confidence_{confidence_method}"]
accuracies = (filtered_df[f"grader_result_{confidence_method}"] == 'A').astype(int)
ece_score_value = ece_score(accuracies, confidence_scores)
auroc_score_value = roc_auc_score(accuracies, confidence_scores)
print(f"confidence_method: {confidence_method}, ECE score: {ece_score_value:.4f}, AUROC score: {auroc_score_value:.4f}, number of samples: {len(filtered_df)}/{len(df)}")
```

```python
df['confidence_vanilla_not_attempted_as_0'] = df['confidence_vanilla'].where(df['grader_result_vanilla'] != 'C', 0)
df['confidence_vanilla_uncertainty_not_attempted_as_0'] = df['confidence_vanilla_uncertainty'].where(df['grader_result_vanilla_uncertainty'] != 'C', 0)
df['confidence_verbal_numerical_confidence_not_attempted_as_0'] = df['confidence_verbal_numerical_confidence'].where(df['grader_result_verbal_numerical_confidence'] != 'C', 0)
df['confidence_verbal_numerical_confidence_our_mapper_not_attempted_as_0'] = df['confidence_verbal_numerical_confidence_our_mapper'].where(df['grader_result_verbal_numerical_confidence'] != 'C', 0)

df.head()
```

```python
confidence_method = "verbal_numerical_confidence"

import matplotlib.pyplot as plt
from metrics import ECE
from sklearn.metrics import roc_auc_score

ece_score = ECE(n_bins=15)

confidence_scores = df[f"confidence_{confidence_method}_our_mapper_not_attempted_as_0"]
accuracies = (df[f"grader_result_{confidence_method}"] == 'A').astype(int)
ece_score_value = ece_score(accuracies, confidence_scores)
auroc_score_value = roc_auc_score(accuracies, confidence_scores)
print(f"confidence_method: {confidence_method}, ECE score: {ece_score_value:.4f}, AUROC score: {auroc_score_value:.4f}, number of samples: {len(filtered_df)}/{len(df)}")
```
